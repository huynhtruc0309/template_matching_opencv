```python
import cv2
import utils
from pathlib import Path
import random
from MTM import drawBoxesOnRGB
import json
import numpy as np
```

```python
def ioa(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
 
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)
    boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
    
    return interArea / float(boxAArea)

def calc_acc(json_data, boxes, num_template, intersect=0.8):
    acc = 0.0
    for shape in json_data['shapes']:
        for box in list(boxes[boxes['TemplateName'] == shape['label']].BBox):
            points = np.asarray(shape['points'], dtype=int)
            x1, y1 = points.min(axis=0)
            x2, y2 = points.max(axis=0)
                        
            (x, y, w, h) = box
            if ioa((x1, y1, x2, y2), (x, y, x + w, y + h)) > intersect:
                acc += 1.0
                continue
    return acc / float(num_template)
```

```python
config = utils.load_yaml('./modules/multi_template_matching/config.yaml')
matcher = utils.create_instance(config['mtm_opencv'])
```

```python
image_dir = Path('/media/huynhtruc0309/DATA/vtcc/CPP/TEMPLATE_MATCHING/GD2/1. CPP/85_Switzerland_equal/total/1_CPP_Page1')
output_dir = Path('output')
image_paths = list(image_dir.glob('**/*.jpg'))
random.shuffle(image_paths)
print(len(image_paths))
```

```python
split_heights, split_widths = [1,], [5]
score_threses = [0.3488372093023256]

bestSplit = None

for score_thres in score_threses:
    for split_height in split_heights:
        for split_width in split_widths:
            print('**' * 20)
            print('score_threses', 'split_heights', 'split_widths', score_thres, split_height, split_width)
            acc = 0.0
            for idx, image_path in enumerate(image_paths):
            #     if idx == 10:
            #         break
                print(image_path)
#                 print('**' * 20)
                image = cv2.imread(str(image_path))
                Hit = matcher(image, split_height, split_width, score_thres)
                print(Hit)

                json_pth = str(image_path).replace('jpg', 'json')
                with open(json_pth) as json_file:
                    json_data = json.load(json_file)
                    acc_per_image = calc_acc(json_data, Hit, len(matcher.listTemplate), intersect=0.8)
                    print(acc_per_image)
                    acc += acc_per_image

#                 Overlay = drawBoxesOnRGB(image, Hit, showLabel=True, boxColor=(0, 0, 0), labelColor=(0, 0, 0), labelScale=2 )
#                 cv2.imshow('Overlay', cv2.resize(Overlay, (0,0), fx=0.5, fy=0.5))
#                 cv2.waitKey()
#                 cv2.destroyAllWindows()

            acc /= float(len(image_paths))
        print(acc)
        if bestSplit == None or acc > bestSplit[1]:
            bestSplit = (split_height, split_width, acc)

        print('bestTemplate', bestSplit) # bestTemplate (1, 5, 0.3488372093023256)
```
